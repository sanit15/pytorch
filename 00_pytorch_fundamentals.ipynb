`Start with random numbers -> look at data -> update random numbers -> repeat `

## Reshaping, Stacking, Squeezing and unsqueezing

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view oif the input tensorof a certain shape but keep the same memory
* Stacking - combines multi tensors on top of each other or side by side
* Squeeze - removes all `1` dim target tensors
* Unsqueeze - adds a `1` dim to a target
* permute - return a view of the input with dimensions permuted in a certain way

```python
#lets create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
x_reshape = x.reshape(1,9)
x_reshape, x_reshape.shape
```

```python
x_reshape = x.reshape(9,1)
x_reshape, x_reshape.shape
```

```python
#change the view
z = x.view(1,9)
z, z.shape
```

```python
#hence changing z changes x
z[:, 0] = 5
z,x
```

```python
#stack tensors on top of each other
x_stack = torch.stack([x,x,x,x], dim = 1)
x_stack
```

```python
x_squeeze = torch.squeeze(x_reshape)
x_squeeze, x_squeeze.shape ## removes dim1 and hence 1,9 -> 9
```

```python
#torch.unsqueze() - adds a single dim to target at specific dim
x_unsqueezed = torch.unsqueeze(x_squeeze, dim=0)
x_unsqueezed, x_unsqueezed.shape
#
```

```python
#torch.permute()
x_original = torch.rand(size=(2,2,3)) # height, width, color_channel
x_permuted = x_original.permute(2,0,1) # color_channel, height, width
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
x_original, x_permuted
```

##Reproducability
in short how a neural network works
`start with random numbers -> tensor operations -> update random numbers to try and make them better rpresentation of the date -> repeat`

To reduce randomness in neural network we use of concept of a **random seed**

```python
##Reproducibility(trying to take random out of random)
rand1 = torch.rand(3,4)
rand2 = torch.rand(3,4)
print(rand1)
print(rand2)
print(rand1 == rand2)
```

```python
#lets make a random but reproducable tensor
import torch
#set random seed
rand_seed = 42
torch.manual_seed(rand_seed)
rand1 = torch.rand(3,4)
torch.manual_seed(rand_seed)
rand2 = torch.rand(3,4)
print(rand1)
print(rand2)
print(rand1 == rand2)
```
